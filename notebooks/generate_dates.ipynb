```python
import pandas as pd
import matplotlib.pyplot as plt
%cd ..
from src.data_loader import OccupancyData
%cd -
```

```python
data = OccupancyData("../data/hospitalization.csv")
icu = data.get_df("2020-11-10", "2022-09-30", "Österreich", "ICU")
normal = data.get_df("2020-11-10", "2022-09-30", "Österreich", "normal")
```

```python
N = 150
dates = pd.date_range(start='2020-12-15', end='2022-09-15', periods=N)
dates = [date.date() for date in dates]
plt.show()
plt.plot(icu)
plt.plot(icu.loc[dates], linestyle="", marker="o", markersize=4)
plt.title("ICU occupancy")
plt.show()
plt.plot(normal)
plt.plot(normal.loc[dates], linestyle="", marker="o", markersize=4)
plt.title("Non-ICU occupancy")
plt.show()
for date in dates:
    print(f'- "{date}"')
```
